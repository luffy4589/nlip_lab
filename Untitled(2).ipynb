```python
sp_tagged = []

for sent in tagged_sentences:
    words = [w for w, t in sent]
    tags = [t for w, t in sent]

    sp_words = []
    sp_tags = []

    for word, tag in zip(words, tags):
        pieces = sp.encode_as_pieces(word)

        for p in pieces:
            sp_words.append(p)
            sp_tags.append(tag)

    sp_tagged.append(list(zip(sp_words, sp_tags)))
```

```python
from nltk.tag import hmm

trainer = hmm.HiddenMarkovModelTrainer()

hmm_tagger = trainer.train_supervised(train_data)

print("HMM Trained with SentencePiece Tokens!")
```

```python
import pandas as pd

transitions = hmm_tagger._transitions

rows = []

# Correct way to iterate through ConditionalProbDist
for prev_tag in transitions.conditions():
    dist = transitions[prev_tag]
    for next_tag in dist.samples():
        prob = dist.prob(next_tag)
        rows.append([prev_tag, next_tag, prob])

trans_table = pd.DataFrame(rows, columns=["Previous Tag", "Next Tag", "Probability"])

print("\n--- TRANSITION PROBABILITY TABLE ---")
print(trans_table.head(50))
```

```python
import pandas as pd

print("\n--- EMISSION PROBABILITY TABLE ---")

# Access emission probabilities from HMM tagger
emission_dict = hmm_tagger._outputs

rows = []

for tag in emission_dict.conditions():
    for word in emission_dict[tag].samples():
        prob = emission_dict[tag].prob(word)
        rows.append([tag, word, prob])

emission_table = pd.DataFrame(rows, columns=["POS Tag", "Word", "Probability"])

# Sort for better readability
emission_table = emission_table.sort_values(by=["POS Tag", "Probability"], ascending=[True, False])
clean_emission = emission_table[~emission_table["Word"].isin(["▁", ",", ".", "-", "$"])]
print(clean_emission.head(30))
```

```python
def viterbi(sentence, states, start_p, trans_p, emit_p):
    V = [{}]
    path = {}

    # Initialization step
    for state in states:
        start_prob = start_p.prob(state) if state in start_p.samples() else 1e-6
        emit_prob = emit_p[state].prob(sentence[0]) if sentence[0] in emit_p[state].samples() else 1e-6

        V[0][state] = start_prob * emit_prob
        path[state] = [state]

    # Recursion step
    for t in range(1, len(sentence)):
        V.append({})
        newpath = {}

        for curr_state in states:
            max_prob = -1
            best_prev = None

            for prev_state in states:
                trans_prob = trans_p[prev_state].prob(curr_state) if curr_state in trans_p[prev_state].samples() else 1e-6
                emit_prob = emit_p[curr_state].prob(sentence[t]) if sentence[t] in emit_p[curr_state].samples() else 1e-6

                prob = V[t - 1][prev_state] * trans_prob * emit_prob

                if prob > max_prob:
                    max_prob = prob
                    best_prev = prev_state

            V[t][curr_state] = max_prob
            newpath[curr_state] = path[best_prev] + [curr_state]

        path = newpath

    # Termination step
    max_prob = -1
    best_final = None

    for state in states:
        if V[-1][state] > max_prob:
            max_prob = V[-1][state]
            best_final = state

    return path[best_final], V
```

# Part 2 : POS Tagging using Conditional Random Fields (CRF)

```python
# -------- FEATURE ENGINEERING FOR CRF --------

def word2features(sent, i):
    word = sent[i]

    features = {
        'bias': 1.0,
        'word.lower()': word.lower(),
        'word.isupper()': word.isupper(),
        'word.istitle()': word.istitle(),
        'word.isdigit()': word.isdigit(),
        'suffix-3': word[-3:],
        'suffix-2': word[-2:],
        'prefix-3': word[:3],
        'prefix-2': word[:2],
        'word.len': len(word),
        'has-hyphen': '-' in word,
        'is_capitalized': word[0].isupper()
    }

    if i > 0:
        prev_word = sent[i-1]
        features.update({
            '-1:word.lower()': prev_word.lower(),
            '-1:isupper()': prev_word.isupper(),
            '-1:istitle()': prev_word.istitle(),
        })
    else:
        features['BOS'] = True   # Beginning of sentence

    if i < len(sent)-1:
        next_word = sent[i+1]
        features.update({
            '+1:word.lower()': next_word.lower(),
            '+1:isupper()': next_word.isupper(),
            '+1:istitle()': next_word.istitle(),
        })
    else:
        features['EOS'] = True   # End of sentence

    return features


def sent2features(sent):
    return [word2features([w for w, t in sent], i) for i in range(len(sent))]

def sent2labels(sent):
    return [tag for word, tag in sent]

def sent2tokens(sent):
    return [word for word, tag in sent]
```

```python
X_train = [sent2features(s) for s in train_data]
y_train = [sent2labels(s) for s in train_data]

X_test = [sent2features(s) for s in test_data]
y_test = [sent2labels(s) for s in test_data]

print("Training samples:", len(X_train))
print("Test samples:", len(X_test))
```

```python
import sklearn_crfsuite
from sklearn_crfsuite import metrics

crf = sklearn_crfsuite.CRF(
    algorithm='lbfgs',
    max_iterations=100,
    all_possible_transitions=True
)

print("Training CRF model...")
crf.fit(X_train, y_train)
print("Training completed!")
```

```python
y_pred = crf.predict(X_test)

print("Sample Prediction:")
print("Words :", sent2tokens(test_data[0]))
print("True  :", y_test[0])
print("Pred  :", y_pred[0])
```

```python
labels = list(crf.classes_)

print("\n================ CRF PERFORMANCE =================")
print(metrics.flat_classification_report(
    y_test, y_pred, labels=labels, digits=3
))
```

```python
from collections import Counter

print("\nTop Positive Features:")

for (attr, label), weight in Counter(crf.state_features_).most_common(20):
    print(f"{weight:.3f} {attr} → {label}")

print("\nTop Negative Features:")

for (attr, label), weight in Counter(crf.state_features_).most_common()[-20:]:
    print(f"{weight:.3f} {attr} → {label}")
```

# Part 3 : Named Entity Recognition using HMM and CRF

```python
import spacy

print("Loading Swedish spaCy model...")
nlp = spacy.load("sv_core_news_sm")
print("Model loaded!")

sentences = []
ner_tags = []

print("Generating BIO tags from captions...")

for i, caption in enumerate(df["caption"]):

    doc = nlp(str(caption))
    
    words = []
    tags = []
    
    for token in doc:
        words.append(token.text)
        
        if token.ent_iob_ == "O":
            tags.append("O")
        else:
            tags.append(token.ent_iob_ + "-" + token.ent_type_)
    
    sentences.append(words)
    ner_tags.append(tags)

    if i % 5000 == 0:
        print(f"Processed {i} captions...")

print("\nDone!")
print("Total sentences:", len(sentences))
```

## PART B – CRF BASED NER

```python
def ner_features(sent, i):

    word = sent[i]

    features = {
        'word': word,
        'lowercase': word.lower(),
        'is_capitalized': word[0].isupper(),
        'is_digit': word.isdigit(),
        'suffix-3': word[-3:],
        'prefix-3': word[:3],
    }

    if i > 0:
        features['prev_word'] = sent[i-1]
    else:
        features['BOS'] = True

    if i < len(sent)-1:
        features['next_word'] = sent[i+1]
    else:
        features['EOS'] = True

    return features
```

```python
X_train = [[ner_features(s, i) for i in range(len(s))] for s in train_sents]
y_train = train_tags

X_test = [[ner_features(s, i) for i in range(len(s))] for s in test_sents]
y_test = test_tags
```

```python
from sklearn_crfsuite import CRF

print("Training CRF model...")

crf = CRF()
crf.fit(X_train, y_train)

print("CRF Training Complete!")
```

```python
y_pred_crf = crf.predict(X_test)
print(y_pred_crf)
```

```python
from sklearn_crfsuite import metrics

crf_acc = metrics.flat_accuracy_score(y_test, y_pred_crf)

print("\nCRF NER Accuracy:", crf_acc)
```

# Part 4 : POS Tagging using Neural Models (BiLSTM)

```python
import spacy
import pandas as pd

# Load Swedish model
print("Loading Swedish model...")
nlp = spacy.load("sv_core_news_sm")
print("Model loaded!")

sentences = []
pos_tags = []

print("Generating POS tags...")

for i, caption in enumerate(df["caption"]):

    doc = nlp(str(caption))

    words = []
    tags = []

    for token in doc:
        words.append(token.text)
        tags.append(token.pos_)      # Universal POS tag

    sentences.append(words)
    pos_tags.append(tags)

    if i % 5000 == 0:
        print(f"Processed {i} captions...")

print("\nDone!")
print("Total sentences:", len(sentences))
```

```python
from sklearn.model_selection import train_test_split

train_sentences, test_sentences, train_pos, test_pos = train_test_split(
    sentences,
    pos_tags,
    test_size=0.2,
    random_state=42
)

print("Training size:", len(train_sentences))
print("Test size:", len(test_sentences))

train_pos = [[tag.upper() for tag in sent] for sent in train_pos]
test_pos  = [[tag.upper() for tag in sent] for sent in test_pos]

```

```python
print("Sample sentence:", train_sentences[0])
print("Sample POS tags:", train_pos[0])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import to_categorical
import numpy as np

# Word tokenizer
word_tokenizer = Tokenizer(oov_token="<OOV>")
word_tokenizer.fit_on_texts(train_sentences)

X_train = word_tokenizer.texts_to_sequences(train_sentences)
X_test  = word_tokenizer.texts_to_sequences(test_sentences)

# Ensure POS tags are uppercase
train_pos = [[tag.upper() for tag in sent] for sent in train_pos]
test_pos  = [[tag.upper() for tag in sent] for sent in test_pos]

# POS tag tokenizer
tag_tokenizer = Tokenizer(lower=True)
tag_tokenizer.fit_on_texts(train_pos)

y_train = tag_tokenizer.texts_to_sequences(train_pos)
y_test  = tag_tokenizer.texts_to_sequences(test_pos)

# Use realistic max length
MAX_LEN = max(len(s) for s in train_sentences)

X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding="post")
X_test  = pad_sequences(X_test, maxlen=MAX_LEN, padding="post")

y_train = pad_sequences(y_train, maxlen=MAX_LEN, padding="post")
y_test  = pad_sequences(y_test, maxlen=MAX_LEN, padding="post")

y_train = to_categorical(y_train, num_classes=len(tag_tokenizer.word_index)+1)
y_test  = to_categorical(y_test, num_classes=len(tag_tokenizer.word_index)+1)

num_words = len(word_tokenizer.word_index) + 1
num_tags  = len(tag_tokenizer.word_index) + 1

print("Vocabulary size:", num_words)
print("Number of POS tags:", num_tags)
print("Max sequence length:", MAX_LEN)
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
print("Loading FastText embeddings...")

embedding_index = {}

with open("cc.sv.300.vec/cc.sv.300.vec", encoding="utf-8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype="float32")
        embedding_index[word] = vector

print("Total embeddings:", len(embedding_index))
```

```python
embedding_dim = 300

embedding_matrix = np.zeros((num_words, embedding_dim))

for word, i in word_tokenizer.word_index.items():
    vector = embedding_index.get(word)
    if vector is not None:
        embedding_matrix[i] = vector
```

```python
model = Sequential([
    Embedding(
        input_dim=num_words,
        output_dim=300,
        input_length=MAX_LEN,
        trainable=True,
        mask_zero=True
    ),
    Bidirectional(LSTM(128, return_sequences=True)),
    TimeDistributed(Dense(num_tags, activation="softmax"))
])

model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python


history = model.fit(
    X_train, y_train,
    validation_data=(X_test, y_test),
    epochs=5,
    batch_size=32
)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history["accuracy"], label="Train Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.title("BiLSTM POS Tagger Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("BiLSTM POS Tagger Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
